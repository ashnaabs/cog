## Exploratory data analysis

```python
df=df.drop(['Unnamed: 32','id'],axis=1)
```

```python
df.diagnosis.replace({'M' : '1','B': '0'},inplace=True)
```

```python
df['diagnosis']=df['diagnosis'].astype(int)
```

```python
cor=df.corr()
```

```python
plt.figure(figsize=(25,23))
sns.heatmap(cor, annot= True, linewidths= 0.3 ,linecolor = "black", fmt = ".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
threshold = 0.75
filtre = np.abs(cor['diagnosis'] > threshold)
corr_features = cor.columns[filtre].tolist()
```

```python
#graph showingcorrelation greater than 0.75
sns.pairplot(df[corr_features], diag_kind = "kde" , markers = "*", hue="diagnosis")
plt.show()
```

```python
sns.countplot(x=df['diagnosis'],data=df)
```

## Data splitting

```python
x=df.iloc[:,1:].values
y=df.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.30, random_state=42)
```

## Model creation

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay,classification_report
a1=(accuracy_score(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
random=RandomForestClassifier()
random.fit(x_train,y_train)
y_pred1=random.predict(x_test)
```

```python
a2=(accuracy_score(y_test,y_pred1))
print(accuracy_score(y_test,y_pred1))
print(classification_report(y_test,y_pred1))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred1))
```

```python
from sklearn.naive_bayes import GaussianNB
bayes=GaussianNB()
bayes.fit(x_train,y_train)
y_pred2=bayes.predict(x_test)
```

```python
a3=(accuracy_score(y_test,y_pred2))
print(accuracy_score(y_test,y_pred2))
print(classification_report(y_test,y_pred2))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred2))
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression(random_state=42)
log.fit(x_train,y_train)
y_pred3=log.predict(x_test)
```

```python
a4=accuracy_score(y_test,y_pred3)
print(accuracy_score(y_test,y_pred3))
print(classification_report(y_test,y_pred3))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred3))
```

## Conclusion and comparison

```python
models=['KNeighbors','Random forest','Naive bayes','Logistic regression']
accuracy=[a1,a2,a3,a4]
fig = plt.figure(figsize=(20,10))
plt.plot(models,accuracy,label='Accuracy',lw=5,color='peru',marker='o',markersize = 15)
plt.legend(fontsize=15)
plt.xlabel('model',fontsize= 20)
plt.ylabel('Accuracy',fontsize= 20)
plt.show()
```
